```python
def rk4solve(f, x0, y0, xmax, h):
    
    Y = np.array([y0])
    X = np.arange(x0, xmax, h)
    
    for x in X[:-1]:
    
        y = Y[-1] # current y_value

        # append y(x+h)
        k1 = f(x, y)
        k2 = f(x+1/2*h, y+1/2*k1*h)
        k3 = f(x+1/2*h, y+1/2*k2*h)
        k4 = f(x + h, y + k3)

        Y = np.append(Y, y+h/6*(k1 + 2*k2 + 2*k3 + k4))
        
    
    return (X, Y)
```

```python
def f(x, y):
    #fout = x*(1+4*y**2)
    f0 = y[1]
    f1 = -4*y[0]
    fout = np.array([f0,f1])
    return fout

# Apply Runge-Katta Method
def rk4solve(f,x0,y0,xmax,h):
    xvals = np.array([x0])
    yvals = np.array([y0])
    
    x = x0
    y = y0
    while x<xmax:
        
        k1 = h*f(x,y)
        k2 = h*f(x+0.5*h,y+0.5*k1)
        k3 = h*f(x+0.5*h,y+0.5*k2)
        k4 = h*f(x+h,y+k3)
        
        yn = y + k1/6.0 + k2/3.0 +k3/3.0 + k4/6.0
        
        xn = x + h
    
        xvals = np.vstack([xvals,xn])
        yvals = np.vstack([yvals,yn])
        x,y = xn,yn
        #print(xn,yn)
        
    return xvals,yvals
    
    
x0 = 0
y0 = np.array([1,0])
h = 0.01
xmax = 4.0
xvals,yvals = rk4solve(f,x0,y0,xmax,h)
plt.plot(xvals,yvals)
```

```python
def rk4solve(fprime, x0, xmax, y0):

    h = 10**-4
    num_points = ceil((xmax-x0)/h)
    X = np.linspace(x0, xmax, num_points) # Store x-values
    Y = np.array(y0, ndmin =2) # Store y-values
    
    for x in X[:-1]:

        y = Y[-1] # use current y-value i.e. y and current x-value i.e. x

        k1 = fprime(x, y)
        k2 = fprime(x+1/2*h, y+1/2*k1*h)
        k3 = fprime(x+1/2*h, y+1/2*k2*h)
        k4 = fprime(x + h, y + k3*h)

        Y = np.append(Y, [y+h/6*(k1 + 2*k2 + 2*k3 + k4)], axis=0)
        
    return (X, Y)
```
