```python
from math import *
import numpy as np
import myquad as myq
import matplotlib.pyplot as plt
```

```python
def squad2d(f, a, b, c, d):
    
    def yint(f, x, c, d):
        
        f1 = lambda y: f(x, y)
        yint = myq.quadz(f1, c, d, 1e-4)
        
        return yint
        
    f2 = lambda x: yint(f, x, c, d)
    
    squad2d = myq.quadz(f2, a, b, 1e-4)
    
    
    
    return squad2d
```

```python
def f(x, y):
    return (x**2)*y

squad2d(f, 0, 1, 0, 2)
```

```python
def region(x, y):
    
    arr = x**2 + y**2
    
    bool_mask = np.where(arr <1)
    
    return bool_mask
```

```python
bool_mask = region(x, y)

x = p[bool_mask, 0]
y = p[bool_mask, 1]

plt.scatter(x, y)
```
